```python
# Preprocessing colors:

df[['R', 'G', 'B']] = df['des_color_specification_esp'].map(colores_rgb).apply(pd.Series)

df.drop(["des_color_specification_esp", "cod_color_code", "des_agrup_color_eng"], axis=1, inplace=True)
```

```python
def one_hot(df, var):
    # Aplicar la codificación one-hot
    one_hot_encoded = pd.get_dummies(df[var], prefix=var)

    # Concatenar las columnas codificadas al DataFrame original
    df_encoded = pd.concat([df, one_hot_encoded], axis=1)

    df_encoded.drop(var, axis=1, inplace=True)

    return df_encoded
```

```python
def prepr(df, par):
    print("Cat: ", df[par].unique())

    print(len(df[par].unique()))

    print(df[par].value_counts())
```

```python
def eliminar_filas_por_categoria(dataframe, variable_categorica, categorias_a_eliminar):
    """
    Elimina las filas del DataFrame basándose en una variable categórica y ciertas categorías específicas.
    Agrega la variable 'cod_modelo_color' de las filas eliminadas a la variable global PROD_ELIMINATS.

    Parámetros:
    - dataframe: DataFrame de pandas.
    - variable_categorica: Nombre de la columna que contiene la variable categórica.
    - categorias_a_eliminar: Lista de categorías que se deben eliminar.

    Devuelve:
    - Nuevo DataFrame sin las filas que contienen las categorías especificadas.
    """
    # Verificar si la variable categórica existe en el DataFrame
    if variable_categorica not in dataframe.columns:
        raise ValueError(f"La variable categórica '{variable_categorica}' no existe en el DataFrame.")

    # Filtrar las filas que no pertenecen a las categorías a eliminar
    filas_eliminadas = dataframe[dataframe[variable_categorica].isin(categorias_a_eliminar)].copy()

    # Agregar 'cod_modelo_color' de las filas eliminadas a la variable global
    global PROD_ELIMINATS
    PROD_ELIMINATS.extend(filas_eliminadas['cod_modelo_color'].tolist())

    # Crear un nuevo DataFrame sin las filas que contienen las categorías especificadas
    nuevo_dataframe = dataframe[~dataframe[variable_categorica].isin(categorias_a_eliminar)].copy()

    return nuevo_dataframe
# Ejemplo de uso:
# Supongamos que tienes un DataFrame llamado df y deseas eliminar las filas con la variable categórica 'Categoria' que contiene las categorías 'A' y 'B'.
# df = eliminar_filas_por_categoria(df, 'Categoria', ['A', 'B'])
```

```python
from scipy.stats import chi2_contingency

df2 = df.drop(["cod_modelo_color", "cod_color_code", "des_color_specification_esp", "des_agrup_color_eng", "des_filename"], axis=1)
l = set()
# Realizar la prueba de chi-cuadrado para cada par de variables categóricas
for col1 in df2.columns:
    for col2 in df2.columns:
        if col1 != col2 and (col1,col2) not in l and (col2, col1) not in l:
            chi2, p, _, _ = chi2_contingency(pd.crosstab(df[col1], df[col2]))
            l.add((col1, col2))
            l.add((col2, col1))
            if p <= 0.05 and p != 0:
                print(f"Correlación entre '{col1}' y '{col2}': Chi2 = {chi2}, p-value = {p}")
```
